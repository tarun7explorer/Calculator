```python
from IPython.display import display, HTML

html_code = """
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Enhanced Calculator</title>
    <style>
      body { font-family: Arial, sans-serif; margin: 40px; background: #f5f5f5;}
      .calculator { background: #fff; padding: 20px 40px 40px 40px; border-radius: 8px; max-width: 400px; margin: auto; box-shadow: 0 4px 20px rgba(0,0,0,0.1);}
      input[type="number"] { width: 160px; margin: 5px;}
      button { margin: 6px 3px 0 0; padding: 10px 18px; font-size: 1em; border: 1px solid #ddd; border-radius: 5px; background: #eee; cursor: pointer;}
      button:hover { background: #ddd;}
      h2, h3 { text-align: center;}
    </style>
  </head>
  <body>
    <div class="calculator">
      <h2>Simple Calculator</h2>
      <input type="number" id="num1" placeholder="Enter number 1" />
      <input type="number" id="num2" placeholder="Enter number 2" />
      <br /><br />
      <button onclick="calculate('+')">+</button>
      <button onclick="calculate('-')">-</button>
      <button onclick="calculate('*')">*</button>
      <button onclick="calculate('/')">/</button>
      <button onclick="calculate('^')">x<sup>y</sup></button>
      <button onclick="calculate('%')">%</button>
      <br /><br />
      <button onclick="calculate('sqrt1')">√Num 1</button>
      <button onclick="calculate('sqrt2')">√Num 2</button>
      <button onclick="changeSign('num1')">+/- Num 1</button>
      <button onclick="changeSign('num2')">+/- Num 2</button>
      <br /><br />
      <button onclick="clearCalculator()">Clear</button>
      <br /><br />
      <h3>Result: <span id="result"></span></h3>
    </div>
    <script>
      function calculate(operation) {
        const num1 = parseFloat(document.getElementById("num1").value);
        const num2 = parseFloat(document.getElementById("num2").value);
        let result;
        if (["+", "-", "*", "/", "^", "%"].includes(operation)) {
          if (isNaN(num1) || isNaN(num2)) {
            alert('Please enter valid numbers!');
            return;
          }
        }
        switch (operation) {
          case '+': result = num1 + num2; break;
          case '-': result = num1 - num2; break;
          case '*': result = num1 * num2; break;
          case '/':
            if (num2 === 0) result = "Division by zero!";
            else result = num1 / num2;
            break;
          case '^': result = Math.pow(num1, num2); break;
          case '%': result = (num1 / 100) * num2; break;
          case 'sqrt1':
            if (isNaN(num1)) { alert('Enter valid number in Num 1!'); return;}
            result = Math.sqrt(num1); break;
          case 'sqrt2':
            if (isNaN(num2)) { alert('Enter valid number in Num 2!'); return;}
            result = Math.sqrt(num2); break;
          default: result = 'Invalid operation';
        }
        document.getElementById("result").innerText = result;
      }
      function clearCalculator() {
        document.getElementById("num1").value = '';
        document.getElementById("num2").value = '';
        document.getElementById("result").innerText = '';
      }
      function changeSign(inputId) {
        const input = document.getElementById(inputId);
        let value = parseFloat(input.value);
        if (isNaN(value)) {
          alert('Please enter a valid number!');
          return;
        }
        value = -value;
        input.value = value;
      }
    </script>
  </body>
</html>
"""

display(HTML(html_code))
```
